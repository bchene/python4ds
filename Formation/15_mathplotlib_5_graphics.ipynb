```python
import matplotlib.pyplot as plt
import numpy as np
```

### 5 - Classification : plt.scatter

```python
from sklearn.datasets import load_iris  # importation de la dataset iris
iris = load_iris()
iris_x = iris.data
iris_y = iris.target
names_iris = list(iris.target_names)

print(f'iris_x contient {iris_x.shape[0]} exemples et {iris_x.shape[1]} variables')
print(f'Il y a {np.unique(iris_y).size} classes : {names_iris}')
```

```python
plt.scatter(iris_x[:, 0], iris_x[:, 1], c=iris_y, alpha=0.5, s=iris_x[:,2]*100)
plt.xlabel("longueur spale")
plt.ylabel("largeur spale")
plt.title('iris model classification')
plt.show()
```

### 4 - Graphique 3d : npl_toolkit.mplot3d  POO

```python
# %matplotlib si interactif
# %matplotlib widget
# ou
# Backend statique (non interactif) pour graphiques 3D
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

ax = plt.axes(projection='3d')
ax.scatter(iris_x[:, 0], iris_x[:, 1], iris_x[:, 2], c=iris_y, alpha=0.5)
ax.set_xlabel('Longueur spale')
ax.set_ylabel('Largeur spale')
ax.set_zlabel('Longueur ptale')
ax.set_title('Graphique 3D - Iris Dataset')
plt.show()
```

```python
# fonction anonyme f(x,y) = sin(x) + cos(x + y)
f = lambda x, y: np.sin(x) + np.cos(x + y)
fx = np.linspace(0, 5, 100)
fy = np.linspace(0, 5, 100)
fx, fy = np.meshgrid(fx, fy)
fz = f(fx, fy)

ax = plt.axes(projection='3d')
ax.plot_surface(fx, fy, fz, cmap='viridis')  # cmap = colormap
plt.show()
```

### 3 - Histogramme : plt.hist

```python
plt.hist(iris_x[:, 0], bins=20)
plt.hist(iris_x[:, 1], bins=20)
plt.show()
```

```python
plt.hist2d(iris_x[:, 0], iris_x[:, 1], cmap='jet')
plt.xlabel("longueur spale")
plt.ylabel("largeur spale")
plt.colorbar()
plt.show()
```

```python
# Histogramme d'image sans les valeurs extrmes (0 et 255)
image = plt.imread('numpy_photo.jpeg')
image_grey = image[:,:,0]
print(image_grey.shape)

# Filtrer les valeurs extrmes (0 et 255) avant de crer l'histogramme
image_filtered = image_grey.flatten()
image_filtered = image_filtered[(image_filtered > 0) & (image_filtered < 255)]

# Crer l'histogramme avec les valeurs filtres
plt.hist(image_filtered, bins=253, range=(1, 254))  # 253 bins pour les valeurs de 1  254
plt.xlabel('Intensit de pixel')
plt.ylabel('Frquence')
plt.title('Histogramme (sans valeurs 0 et 255)')
plt.grid(True, alpha=0.3)
plt.show()
```
