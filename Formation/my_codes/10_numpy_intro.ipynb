```python
import numpy as np
```

### Constructeurs

```python
np.array([1, 2, 3])  # construction d'un array
np.array([1, 2, 3], dtype=float)  # construction d'un array de float
np.array([[1, 2, 3], [4, 5, 6]])  # construction d'un array 2D
np.zeros((3, 2))  # construction d'un array de 0
np.ones((3, 2))  # construction d'un array de 1
np.full((3, 2), 4)  # construction d'un array de 4
np.random.rand(3, 2)  # construction d'un array de nombres aléatoires
np.random.randn(3, 2)  # construction d'un array de nombres aléatoires
np.random.normal(0, 1, (3, 2))  # construction d'un array de nombres aléatoires
np.random.uniform(0, 1, (3, 2))  # construction d'un array de nombres aléatoires
np.eye(3)  # construction d'un array identité
np.linspace(0, 1, 5) # 5 points entre 0 et 1
np.arange(0, 1, 0.1) # de 0 à 1 par pas de 0.1
```

### Affichage de divers Arguments de ndarray
https://numpy.org/devdocs/reference/generated/numpy.ndarray.html

```python
firstArray = np.array([[1, 2, 3], [4, 5, 6]], dtype=int)
print("firstArray :\n", firstArray)
print("firstArray.shape :", firstArray.shape)
print("firstArray.size :", firstArray.size)
print("firstArray.ndim :", firstArray.ndim)
print("firstArray.dtype :", firstArray.dtype)
print("firstArray.itemsize :", firstArray.itemsize)
print("firstArray.nbytes :", firstArray.nbytes)
```

### Methode de manipulation

```python
a = np.zeros((3, 2))
b = np.ones((3, 2))
print("np.vstack((a,b)) or np.concatenate((a,b), axis=0) :\n", np.concatenate((a,b), axis=0))
print("np.hstack((a,b)) or np.concatenate((a,b), axis=1) :\n", np.concatenate((a,b), axis=1))
```

```python
D = np.concatenate((a,b), axis=0)
print("D - D.shape=", D.shape, "D.size=", D.size, "\n", D)
D = D.reshape(3, 4)
print("D.reshape(3, 4) - D.shape=", D.shape, "D.size=", D.size, "\n", D)
```

```python
D = D.squeeze()
D = D.reshape(D.shape[0], 1)
print("D.squeeze() - D.shape=", D.shape, "D.size=", D.size, "\n", D)
E = D.flatten()
print("E.flatten() - E.shape=", E.shape, "E.size=", E.size, "\n", E)
F = D.ravel()
print("F.ravel() - F.shape=", F.shape, "F.size=", F.size, "\n", F)

G = np.array([[1, 2, 3], [4, 5, 6]])
print("G :\n", G)
G.T
print("G.T :\n", G.T)
```
