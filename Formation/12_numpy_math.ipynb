```python
import numpy as np
```

```python
A = np.random.randint(0, 10, (2, 3))
A
```

```python
print("A.sum() =", A.sum())  # somme des éléments
print("A.sum(axis=0) = A.sum(0) =", A.sum(axis=0)) # somme des colonnes
print("A.sum(axis=1) = A.sum(1) =", A.sum(axis=1)) # somme des lignes
print("A.cumsum() =", A.cumsum()) # somme cumulée
print("A.prod() =", A.prod()) # produit des éléments
print("A.cumprod() =", A.cumprod()) # produit cumulé
```

```python
A = np.random.randint(0, 10, (3, 3))
print("A :\n", A)
print("A.min() =", A.min()) # minimum
print("A.max() =", A.max()) # maximum
print("A.argmin(axis=0) =", A.argmin(axis=0), "A.argmin(axis=1) =", A.argmin(axis=1)) # indice du minimum
print("A.argmax(axis=0) =", A.argmax(axis=0), "A.argmax(axis=1) =", A.argmax(axis=1)) # indice du maximum

```

```python
A = np.random.randint(0, 10, (3, 3))
print("A :\n", A)
B = A.copy()
B.sort(axis=0) # tri par colonne
print("B.sort(axis=0) :\n", B)
B.sort(axis=1) # tri par ligne
print("B.sort(axis=1) :\n", B)
C = A.copy()
C.argsort(axis=0) # indice des éléments triés par colonne
print("C.argsort(axis=0) :\n", C)
C = C.argsort(axis=1) # indice des éléments triés par ligne
print("C.argsort(axis=1) :\n", C)
```

### Numpy Statistics functions
https://docs.scipy.org/doc/numpy-1.10.1/reference/routines.statistics.html

```python
print (A)
print ("A.mean() =",A.mean()) # moyenne
print ("A.mean(axis=0) =",A.mean(axis=0)) # moyenne des colonnes
print ("A.mean(axis=1) =",A.mean(axis=1)) # moyenne des lignes
```

```python
print ("A.var() =",A.var()) # variance
```

```python
print ("A.std() =",A.std()) # écart-type
```

```python
print (A)
print ("np.corrcoef(A) =\n",np.corrcoef(A)) # coefficient de corrélation
print ("np.corrcoef(A)[0,1] =",np.corrcoef(A)[0,1]) # coefficient de corrélation
```

```python
print("np.unique(A) =",np.unique(A)) # valeurs uniques
print("np.unique(A, return_counts=True) =",np.unique(A, return_counts=True)) # valeurs uniques et leurs counts
print("np.unique(A, return_index=True) =",np.unique(A, return_index=True)) # valeurs uniques et leurs indices
print("np.unique(A, return_inverse=True) =",np.unique(A, return_inverse=True)) # valeurs uniques et leurs indices inversés
# tri par ordre de fréquence (argsort retourne les indices triés pour que les valeurs soient dans l'ordre croissant)
values, counts = np.unique(A, return_counts=True)
print("values[counts.argsort()] =",values[counts.argsort()]) # valeurs uniques triées par ordre de fréquence
# affichage des valeurs uniques triées par ordre de fréquence
for i, j in zip(values[counts.argsort()], counts[counts.argsort()]):
    print(f"valeur : {i} - apparait : {j} fois")
```

```python
# nan functions for missing values
B = (A.copy()).astype(np.float64)
B[1,1] = np.nan
B[2,2] = np.nan
np.nansum(B) # somme des éléments, en ignorant les NaNs
np.nanprod(B) # produit des éléments, en ignorant les NaNs
np.nanmean(B) # moyenne des éléments, en ignorant les NaNs
np.nanmedian(B) # médiane des éléments, en ignorant les NaNs
np.nanstd(B) # écart-type des éléments, en ignorant les NaNs
np.nanvar(B) # variance des éléments, en ignorant les NaNs

np.isnan(B) # renvoie un tableau de booléens, True si l'élément est NaN, False sinon
np.isnan(B).sum() # nombre de NaNs dans B
np.isnan(B).sum()/B.size # pourcentage de NaNs dans B

B[np.isnan(B)] = 0 # remplacement des NaNs par 0
```

### Numpy Algebre lineaire
https://docs.scipy.org/doc/numpy-1.10.1/reference/routines.linalg.html

```python
A = np.ones(2,3)
B = np.ones(3,2)

A.T # transposition
A.dot(B) # produit matriciel

A = np.random.randint(1, 10, (3, 3))
np.linalg.det(A) # déterminant
np.linalg.inv(A) # inverse

np.linalg.eig(A) # valeurs propres et vecteurs propres
np.linalg.svd(A) # décomposition en valeurs singulières
np.linalg.qr(A) # décomposition QR
```

```python
np.random.seed(0)
A = np.random.randint(0, 100, (10, 5))
A

# normalisation/standardisation (z-score)
mean = A.mean(axis=0) # moyenne des colonnes
std = A.std(axis=0) # écart-type des colonnes
A_std = (A - mean) / std # standardisation

print(A)
print("standardisation (z-score) par colone")
print(A_std)
```
